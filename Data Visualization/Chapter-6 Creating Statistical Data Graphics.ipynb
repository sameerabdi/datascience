### Eyeballing dataset distribution with H
istograms

```python
address= 'C:/Users/samee/Documents/Python For DS Code/Data/mtcars.csv'
cars = pd.read_csv(address)
cars.columns=['car names', 'mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']
cars.index = cars['car names']
mpg= cars['mpg']
cars.head(3)
mpg.plot(kind= 'hist')
plt.xlabel('')
plt.ylabel('milage/gallan')
plt.title('mpg')
plt.legend('mpg')
```

```python
sb.distplot(mpg)
```
