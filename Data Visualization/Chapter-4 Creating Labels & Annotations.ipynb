```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 8,4
```

### Labeling plot features

#### The Functional Method

```python
x = range(1,10)
y = [1,2,3,4,.5,4,3,2,1]
plt.bar(x,y)
plt.grid()
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
z = [1,2,3,4,5]
veh_type = ['bicycle','moter bike', 'car', 'van', 'stroller']
plt.pie(z, labels=veh_type)
plt.show()
```

#### Object Oriented Method

```python
address= 'C:/Users/samee/Documents/Python For DS Code/Data/mtcars.csv'
cars = pd.read_csv(address)
cars.columns=['car names', 'mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']
cars.head(3)
```

```python
mpg = cars.mpg
```

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])
mpg.plot()
ax.set_xticks(range(32))
ax.set_xticklabels(cars['car names'], rotation=60, fontsize='medium')
ax.set_title('Mile/Gallen')
ax.set_xlabel('Car Name')
ax.set_ylabel('Miles/Gallen')
```

### Adding lagend to the Plot

#### The Functional Method

```python
plt.pie(z)
plt.legend(veh_type, loc='best')
plt.show()
```

#### Object Oriented Method

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])
mpg.plot()
ax.legend(loc='best')
ax.set_xticks(range(32))
ax.set_xticklabels(cars['car names'], rotation=60, fontsize='medium')
ax.set_title('Mile/Gallen')
ax.set_xlabel('Car Name')
ax.set_ylabel('Miles/Gallen')
```
